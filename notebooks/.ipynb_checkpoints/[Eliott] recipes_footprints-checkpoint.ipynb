```python
import re
import json
```

```python
with open('../data/Recipes/recipes_jow.json') as json_file:
    jow = json.load(json_file)
```

```python
# Keeping only ingredients
all_ingredients = []

for key in jow:
    for i in jow[key]["ingredients"]:
        try:
            if len(i)==2:
                all_ingredients.append(i[1])
            elif len(i)==3:
                all_ingredients.append(i[2])
        except:
            pass
        
print(len(all_ingredients))

all_ingredients = list(dict.fromkeys(all_ingredients))
print(len(all_ingredients))

all_ingredients = [re.sub("[\(\[].*?[\)\]]", "", x) for x in all_ingredients]
all_ingredients = list(dict.fromkeys(all_ingredients))
print(len(all_ingredients))
```

```python
from textblob import TextBlob
import time
```

```python
english_ingredients = []
```

```python
for i in range(len(all_ingredients)):
    time.sleep(0.5)
    if i >= len(english_ingredients):
        try:
            english_ingredients.append(str(TextBlob(all_ingredients[i]).translate(from_lang = 'fr', to = 'en')))
        except:
            english_ingredients.append(all_ingredients[i])
    else:
        pass
```

```python
df_ingredients = pd.DataFrame(english_ingredients, columns=["jow_ingredients"])
df_ingredients.head()
```

#### Test matching bdd

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def ingredient_check(string, list_ingredients, tresh):
    res = []
    s = string.lower()
    for ingredient in list_ingredients:
        if fuzz.token_sort_ratio(s, ingredient) > tresh:
            res.append(ingredient)
        else:
            pass
    return res

def most_similar_ingredient(string, list_ingredients, tresh):
    res = "no match"
    value = tresh
    s = string.lower()
    for ingredient in list_ingredients:
        if fuzz.token_sort_ratio(s, ingredient) > value:
            res = ingredient
            value = fuzz.token_sort_ratio(s, ingredient)
        else:
            pass
    return res
```

```python
df_ingredients["ademe_ingredients"] = df_ingredients["jow_ingredients"].apply(lambda x: most_similar_ingredient(x, data_ademe["Main ingredient"], tresh=60))
```

```python
# l'unité pour chaque facteur est le gramme (ex: cac = 17 grammes)

dict_equivalence_poids = {"càc" : 5,
    "càs" : 15,
    "ml" : 1,
    "l" : 1000,
    "poignées" : 1,
    "poignée" : 1,
    "pinc." : 1,
    "kg" : 1000,
    "g" : 1,
    "noisette" : 10,
    "cl" : 10,
    "gou." : 7,
    "liv" : 453,
    "portions" : 17.5, #poids d'une vache qui rit (voir bdd)
    "boule" : 30, #poids boule de glace (voir bdd)
    "bou." : 30, #poids botte de basilic ou autre (voir bdd)
    "sac." : 15,
    "tran." : 50, #approximation, à développer en fonction de l'ingrédient
    "cup" : 150, #approximation, à développer en fonction de l'ingrédient
    "cm" : 6 #approximation, focus gingembre (voir bdd)
    }
```

```python
jow_bis = {}
bug = []

for key in jow:

    jow_bis[key] = {}
    jow_bis[key]["cuisson"] = jow[key]["cuisson"]
    jow_bis[key]["ingredients"] = []

    for i in jow[key]["ingredients"]:
        try:

            if len(i)==2: # structure [quantité, ingredient]
                conv = dict_equivalence_poids[i[0].split(" ")[1]]
                
                try:
                    quantite = float(i[0].split(" ")[0])
                except ValueError:
                    num, denom = i[0].split(" ")[0].split('/')
                    quantite = float(num) / float(denom)

                jow_bis[key]["ingredients"].append({"poids": conv*quantite, "ingredient":i[1]})

            elif len(i)==3:

                conv = dict_equivalence_poids[i[1].split(" ")[1]]
                quantite = float(i[1].split(" ")[0])

                jow_bis[key]["ingredients"].append({"poids": conv*quantite, "ingredient":i[2]})

        except : 
            bug.append(i)
```
